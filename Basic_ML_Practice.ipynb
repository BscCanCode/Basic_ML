# ML coding begins 

```python
import sys
import numpy as np
import pandas as pd
import matplotlib
import sklearn
print("== Library versions")
print(f"python :{sys.version.split()[0]}")
print(f"numpy : {np.__version__}")
print(f"pandas :{pd.__version__}")
print(f"matplot:{matplotlib.__version__}")
print(f"sklearn:{sklearn.__version_}")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = [[1],[2],[3],[4]]
y = [2,4,6,8]
model.fit(X,y)
user_value = float(input("Enter the number:"))
prediction = model.predict([[user_value]])
print("The prediction for I/P 5 is:",prediction)
```

```python
import matplotlib.pyplot as plt
X = [[1],[2],[3],[4]]
y = [2,4,6,8]
plt.scatter(X,y,color = 'blue')
plt.legend()
plt.title("Regression model")
plt.xlabel("X")
plt.ylabel("y")
plt.plot(X,model.predict(X),color = 'orange',label = 'RegressionLine')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = [[1],[2],[3],[4],[5]]
y = [30,60,90,120,150]
model.fit(X,y)
user_value = int(input("Enter the input:"))
prediction = model.predict([[user_value]])
print("The predicted value of user input is:",prediction)
```

```python

plt.scatter(X,y,color = 'blue',label="datapoints")
plt.plot(X,model.predict(X),color = 'orange',label = "regression_line")
plt.legend()
plt.xlabel("banana weight")
plt.ylabel("banana prices")
plt.title("Banana weight vs price")
plt.show()
```

```python
plt.scatter(X,model.predict(X),color = 'red')
plt.xlabel("banana weight")
plt.ylabel("banana prices")
plt.title("Banana weight vs price")
plt.show()
```

```python
plt.plot(X,model.predict(X),color = 'black',linestyle = 'dotted')
plt.xlabel("banana weight")
plt.ylabel("banana prices")
plt.title("Banana weight vs price")

plt.show()
```
